```python
mean_value = gamedata['winPlacePerc'].mean()
gamedata['winPlacePerc'].fillna(mean_value, inplace=True)
```

```python
gamedata.isna().sum()
```

```python
plt.figure(figsize=(10, 6))
sns.histplot(gamedata['damageDealt'], kde=True, bins=30)
plt.title('Distribution of Damage Dealt')
plt.xlabel('Damage Dealt')
plt.ylabel('Frequency')
plt.show()

```

```python
# Scatter Plot of 'walkDistance' vs 'kills'
plt.figure(figsize=(10, 6))
sns.scatterplot(x='walkDistance', y='kills', data=gamedata)
plt.title('Walk Distance vs Kills')
plt.xlabel('Walk Distance')
plt.ylabel('Kills')
plt.show()
```

```python
# Bar Plot of 'matchType' vs 'kills'
plt.figure(figsize=(14, 8))
sns.barplot(x='matchType', y='kills', data=gamedata, estimator=sum)
plt.title('Total Kills by Match Type')
plt.xlabel('Match Type')
plt.ylabel('Total Kills')
plt.xticks(rotation=45)
plt.show()
```

```python
# Box Plot of 'matchType' vs 'winPlacePerc'
plt.figure(figsize=(14, 8))
sns.boxplot(x='matchType', y='winPlacePerc', data=gamedata)
plt.title('Win Place Percentage by Match Type')
plt.xlabel('Match Type')
plt.ylabel('Win Place Percentage')
plt.xticks(rotation=45)
plt.show()
```

```python
# Pair Plot
plt.figure(figsize=(12, 12))
sns.pairplot(gamedata[['damageDealt', 'kills', 'walkDistance', 'winPoints']])
plt.suptitle('Pair Plot of Selected Columns', y=1.02)
plt.show()
```

```python
# Count Plot for 'matchType'
plt.figure(figsize=(14, 8))
sns.countplot(data=gamedata, x='matchType')
plt.title('Count of Each Match Type')
plt.xlabel('Match Type')
plt.ylabel('Count')
plt.xticks(rotation=45)
plt.show()
```

```python
# Box Plot for 'kills' across different 'matchType'
plt.figure(figsize=(14, 8))
sns.boxplot(x='matchType', y='kills', data=gamedata)
plt.title('Kills Distribution by Match Type')
plt.xlabel('Match Type')
plt.ylabel('Kills')
plt.xticks(rotation=45)
plt.show()
```

```python

# Line Plot of 'matchDuration' over time (assuming 'matchId' is sorted by time)
plt.figure(figsize=(14, 8))
sns.lineplot(x='matchId', y='matchDuration', data=gamedata.sort_values('matchId'))
plt.title('Match Duration Over Time')
plt.xlabel('Match ID')
plt.ylabel('Match Duration')
plt.xticks(rotation=45)
plt.show()

```

```python
# Histogram of 'winPoints'
plt.figure(figsize=(10, 6))
sns.histplot(gamedata['winPoints'], bins=30, kde=True)
plt.title('Distribution of Win Points')
plt.xlabel('Win Points')
plt.ylabel('Frequency')
plt.show()

```

```python
# Violin Plot of 'boosts' for different 'matchType'
plt.figure(figsize=(14, 8))
sns.violinplot(x='matchType', y='boosts', data=gamedata)
plt.title('Boosts Distribution by Match Type')
plt.xlabel('Match Type')
plt.ylabel('Boosts')
plt.xticks(rotation=45)
plt.show()
```

```python
# Heatmap of Correlation Matrix
plt.figure(figsize=(16, 12))
corr = gamedata.corr()
sns.heatmap(corr, annot=True, fmt='.2f', cmap='coolwarm', linewidths=0.5)
plt.title('Correlation Matrix Heatmap')
plt.show()
```
